```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
unique_isoform_columns = ["window_chr", "window_start", "window_end", "read_id", "read_seq", "sample_of_the_isoform", "read_mapping_start", "read_length"]

isoform_df = pd.read_csv("unique_isoforms.tsv", sep="\t", header=None, names=unique_isoform_columns)
```

```python
isoform_count = isoform_df.groupby(["window_chr", "window_start", "window_end","sample_of_the_isoform"]).count()["read_id"].reset_index()
```

```python
chromosomes = isoform_count["window_chr"].drop_duplicates().to_list()
```

### Need to normalize count by total number of isoforms in each bin

normalized_uniq_isoform_sample_1_bin_10 = # uniq isoform in sample 1 in bin 10 / # total isoforms in sample 1 in bin 10

```python

for i,chrom in enumerate(chromosomes):
#     print(i)
    isoform_count1 = isoform_count[isoform_count["window_chr"] == chrom]
    isoform_count1["unique isoform count"] = isoform_count1["read_id"]
    
    
    fig, ax = plt.subplots(1)
    
    sns.lineplot(data=isoform_count1, x="window_start", y="unique isoform count", hue="sample_of_the_isoform", ax=ax)

    fig.set_size_inches(20,4)
```

```python
a = isoform_count.groupby(["sample_of_the_isoform"]).sum()["read_id"].reset_index()
a["Unique isoform count"] = a["read_id"]
sns.barplot(data=a, x="sample_of_the_isoform", y="Unique isoform count")
plt.title("Average number of unique isoforms in each sample in each chromosome")
```

```python
plt.figure().set_size_inches(10,5)
a = isoform_count.groupby(["window_chr","sample_of_the_isoform"]).sum()["read_id"].reset_index()
a["Unique isoform count"] = a["read_id"]
sns.barplot(data=a, x="window_chr", y="Unique isoform count", hue="sample_of_the_isoform")
plt.title("Average number of unique isoforms in each chromosome across three samples")
```
